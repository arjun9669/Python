```python
reviewer_df = pd.read_csv('/content/reviewer.csv')
```

```python
1. Write a SQL query to find when the movie 'American Beauty' released. Return
movie release year.
```

```python
american_beauty_year = movie_df[movie_df['mov_title'] == 'American Beauty']['mov_year']
print("Release year of 'American Beauty':", american_beauty_year)
```

```python
2. Write a SQL query to find those movies, which were released before 1998. Return
movie title.
```

```python
movies_before_1998 = movie_df[movie_df['mov_year'] < 1998]['mov_title']
print("Movies released before 1998:", movies_before_1998)
```

```python
3. Write a query where it should contain all the data of the movies which were
released after 1995 and their movie duration was greater than 120
```

```python
movies_after_1995_120min = movie_df[(movie_df['mov_year'] > 1995) & (movie_df['mov_time'] > 120)]
print("Movies after 1995 with duration over 120 minutes:", movies_after_1995_120min)
```

```python
4. Write a query to determine the Top 7 movies which were released in United
Kingdom. Sort the data in ascending order of the movie year.
```

```python
movies_in_uk = movie_df[movie_df['mov_rel_country'] == 'United Kingdom'].nlargest(7, 'mov_year')
print("Top 7 movies in the UK:", movies_in_uk)
```

```python
5. Set the language of movie language as 'Chinese' for the movie which has its
existing language as Japanese and the movie year was 2001.
```

```python
movie_df.loc[(movie_df['mov_lang'] == 'Japanese') & (movie_df['mov_year'] == 2001), 'mov_lang'] = 'Chinese'
print("Updated movies to Chinese where language was Japanese and year was 2001:", movie_df)
```

```python
6. Write a SQL query to find name of all the reviewers who rated the movie 'Slumdog Millionaire'
```

```python
slumdog_id = movie_df[movie_df['mov_title'] == 'Slumdog Millionaire']['mov_id']
ratings_for_slumdog = ratings_df[ratings_df['mov_id'].isin(slumdog_id)]
reviewers_for_slumdog = reviewer_df[reviewer_df['rev_id'].isin(ratings_for_slumdog['rev_id'])]
print("Reviewers for Slumdog Millionaire:", reviewers_for_slumdog)
```

```python
7. Write a query which fetch the first name, last name & role played by the actor where output should all exclude Male actors
```

```python
actor_roles = pd.merge(actor_df[actor_df['act_gender'] != 'M'], cast_df, on='act_id')
print("Female actors and their roles:", actor_roles[['act_fname', 'act_lname', 'role']])
```

```python
8. Write a SQL query to find the actors who played a role in the movie 'Annie Hall'.Fetch all the fields of actor table. (Hint: Use the IN operator).
```

```python
annie_hall_id = movie_df[movie_df['mov_title'] == 'Annie Hall']['mov_id']
actors_in_annie_hall = cast_df[cast_df['mov_id'].isin(annie_hall_id)]
annie_hall_actors = actor_df[actor_df['act_id'].isin(actors_in_annie_hall['act_id'])]
print("Actors in Annie Hall:", annie_hall_actors)
```

```python
9. Write a SQL query to find those movies that have been released in countries otherthan the United Kingdom. Return movie title, movie year, movie time, and date of release, releasing country.
```

```python
non_uk_movies = movie_df[movie_df['mov_rel_country'] != 'United Kingdom'][['mov_title', 'mov_year', 'mov_time', 'mov_dt_rel', 'mov_rel_country']]
print("Movies not released in the UK:", non_uk_movies)
```

```python
10. Print genre title, maximum movie duration and the count the number of
movies in each genre. (HINT: By using inner join)
```

```python
genres_movies = pd.merge(movie_genres_df, genres_df, on='gen_id')

genre_movie_details = pd.merge(genres_movies, movie_df, on='mov_id')

genres_movies = pd.merge(movie_genres_df, genres_df, on='gen_id')

genre_movie_details = pd.merge(genres_movies, movie_df, on='mov_id')

genre_stats = genre_movie_details.groupby('gen_title').agg(
    max_duration=pd.NamedAgg(column='mov_time', aggfunc='max'),
    movie_count=pd.NamedAgg(column='mov_id', aggfunc='nunique')
)
```

```python
print(genre_stats.reset_index())
```

```python
12. Write a SQL query to find the movies with the lowest ratings
```

```python
min_rating = ratings_df['rev_stars'].min()
lowest_rated_movies = ratings_df[ratings_df['rev_stars'] == min_rating]
lowest_rated_movies_details = pd.merge(lowest_rated_movies, movie_df, on='mov_id')
print(lowest_rated_movies_details[['mov_title', 'rev_stars']])
```
