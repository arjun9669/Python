```python
10. Add new column as HL_Calories where more than average calories should be categorized as
HIGH and less than average calories should be categorized as LOW
```

```python
average_calories_overall = data['calories'].mean()
def categorize_calories(cal):
    if cal > average_calories_overall:
        return 'HIGH'
    else:
        return 'LOW'

data['HL_Calories'] = data['calories'].apply(categorize_calories)
print(data[['calories', 'HL_Calories']].head())
```
