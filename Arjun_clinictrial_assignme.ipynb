```python
6. Find average age of those were pregnant and not pregnant
```

```python
pregnancy_status = ['Yes', 'No', 'No']
data['Pregnant'] = pregnancy_status
```

```python
average_age_pregnant = data[data['Pregnant'] == 'Yes']['Age'].mean()
average_age_not_pregnant = data[data['Pregnant'] == 'No']['Age'].mean()
print(f"Average age of pregnant: {average_age_pregnant}, not pregnant: {average_age_not_pregnant}")
```

```python
7. Find average blood pressure of those had drug reaction and did not had drug
reaction
```

```python
drug_reaction_status = ['Yes', 'No', 'Yes']
bp_values = [120, 110, 130]
```

```python
data['DrugReaction'] = drug_reaction_status
data['BP'] = bp_values
```

```python
average_bp_reaction = data[data['DrugReaction'] == 'Yes']['BP'].mean()
average_bp_no_reaction = data[data['DrugReaction'] == 'No']['BP'].mean()
print(f"Average BP with drug reaction: {average_bp_reaction}, without drug reaction: {average_bp_no_reaction}")
```

```python
8. Add new column name as ‘Age_group’ and those having age between 16 & 21
should be categorized as Low, more than 21 and less than 35 should be
categorized as Middle and above 35 should be categorized as High.
```

```python
def categorize_age(age):
    if 16 <= age <= 21:
        return 'Low'
    elif 21 < age < 35:
        return 'Middle'
    else:
        return 'High'

data['Age_group'] = data['Age'].apply(categorize_age)
```
